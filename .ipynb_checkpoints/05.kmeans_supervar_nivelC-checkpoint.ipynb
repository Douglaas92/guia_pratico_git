```python
df = pd.read_parquet('base_2024-1_qualif_supervar_nivelC.parquet')
print(f"Quantidade de Linhas:", df.shape[0])
print(f"Quantidade de Colunas:", df.shape[1])
```

```python
if df['CD_UOR'].duplicated().any():
    print("Existem matr칤culas duplicadas no DF")    
else:
    print("Todas as UOR's s칚o 칰nicas, aparecendo uma 칰nica linha no DF")
```

# Vers칚o 1 do Teste

## Gerando uma c칩pia do DF que ser치 usado para modelagem

```python
dropar = ['SEMESTRE', 'CD_UOR', 'CD_PREF_DEP', 'NM_UOR_RDZ', 'CD_UOR_SPER',
          'CD_PREF_SPER', 'NM_RDZ_SPER', 'CD_UOR_GPES', 'CD_PREF_GPES',
          'NM_RDZ_GPES', 'NM_NVL_UOR', 'PREFIXO_GEPES_CARTEIRA', 'CD_CARTEIRA',
          'SG_UF_UOR', 'GRUPO', 'QT_LOT', 'ANOS_EXP_BB']

df_numerico = df.drop(columns=dropar).reset_index(drop=True)
df_numerico.describe()
```

## Padroniza칞칚o dos Dados
- usando StandardScaler() que ir치 manter a m칠dia das vari치veis = 0 e o desvio padr칚o = 1
- isso visa distribuir os valores das vari치veis em uma mesma escala e diminuir a vari칙ncia dos dados 

**Precisei remover a vari치vel "INDICADOR_ANOTACOES" pois ele possu칤 valores infinitos, n칚o permitindo aplica칞칚o da padroniza칞칚o da vari치vel**

```python
df_numerico = df_numerico.drop("INDICADOR_ANOTACOES", axis=1)
obj_std = StandardScaler().fit(df_numerico)
df_obj_std = obj_std.transform(df_numerico)

df_padronizado = pd.DataFrame(df_obj_std, columns=df_numerico.columns, index=df_numerico.index)
df_padronizado.head(1)
```

```python
print("Estrutura do Dataframe Original: Linhas",df.shape[0], "e Colunas",df.shape[1])
print("Estrutura do Dataframe de Teste: Linhas",df_padronizado.shape[0], "e Colunas",df_padronizado.shape[1])
```

## Identifica칞칚o da quantidade de clusters

### M칠todo Elbow para identifica칞칚o do n췈 de clusters

```python
# Elaborado com base na "WCSS": dist칙ncia de cada observa칞칚o para o centroide de seu cluster
# Quanto mais pr칩ximos entre si e do centroide, menores as dist칙ncias internas
# Normalmente, busca-se o "cotovelo", ou seja, o ponto onde a curva "dobra"

elbow = []
K = range(2,21) # ponto de parada pode ser parametrizado manualmente
for k in K:
    kmeanElbow = KMeans(n_clusters=k, init='random', random_state=42).fit(df_padronizado)
    elbow.append(kmeanElbow.inertia_)
    
plt.figure(figsize=(16,8))
plt.plot(K, elbow, marker='o')
plt.xlabel('N췈 Clusters', fontsize=16)
plt.xticks(range(2,21))
plt.ylabel('WCSS', fontsize=16)
plt.title('M칠todo de Elbow', fontsize=16)
plt.show()
```

### Calcular a silhueta para diferentes n칰meros de clusters

```python
silhouette_scores = []
K = range(2, 21)  # Precisei alterar o range, pois a silhueta n칚o 칠 definida para k=1

for k in K:
    kmeans = KMeans(n_clusters=k, init='random', random_state=42).fit(df_padronizado)
    labels = kmeans.labels_
    silhouette_avg = silhouette_score(df_padronizado, labels)
    silhouette_scores.append(silhouette_avg)

# Aqui eu pe칞o para que seja destacado o pico mais alto, que aponta o num칠ro 칍TIMO de clusters
best_k = K[silhouette_scores.index(max(silhouette_scores))]
best_score = max(silhouette_scores)

# Plotar os resultados
plt.figure(figsize=(16, 8))
plt.plot(K, silhouette_scores, marker='o', label='Coeficiente de Silhueta')
plt.axvline(x=best_k, linestyle='--', color='red', label=f'Melhor k: {best_k}')
plt.scatter(best_k, best_score, color='red')  # Destacar o ponto do pico
plt.xlabel('N췈 Clusters', fontsize=16)
plt.xticks(range(2, 21))
plt.ylabel('M칠dia do Coeficiente de Silhueta', fontsize=16)
plt.title('An치lise de Silhueta para Identifica칞칚o do N췈 칍timo de Clusters', fontsize=16)
plt.legend()
plt.show()
```

```python
print(f"Melhor 칤ndice de silhueta apresentou {best_k} clusters com a m칠dia do coeficiente de silhueta = {round(best_score,4)}")
```
