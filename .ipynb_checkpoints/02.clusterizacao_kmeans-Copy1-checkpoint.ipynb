# 2 - **Clusterização**

A técnica de clusterização é uma solução de modelagem de aprendizado **não-supervisionado**, que visa por meio de agrupamento de similaridade dos dados, encontrar padrões para os indicadores/variáveis disponíveis no dataset. A diferença entre os modelos de aprendizado **supervisionados** é que, na criação do modelo de clusterização não usamos a target (labels), inclusive esse é o principal fator para ser nomeado como não-supervisionado. Aqui também não fazemos a separação do dataset entre treino e teste, ou seja, utilizamos todos os dados disponíveis. Por uma boa prática, comumente ocorre que nesse tipo de aprendizado (não-supervisionado) as colunas do dataset são identificadas como **dimensões** e não mais atributos ou variáveis. Seguindo essa premissa, por exemplo, no dataset de Flores (iris dataset) temos 4 dimensões: 

- *altura da pétala*
- *largura da pétala*
- *altura da sépala*
- *largura da sépala*

## K-Means

Importante mencionar que para fins de clusterização, existem vários algoritmos, mas visando trazer um panorama que possa abranger a ideia macro de um cluster, utilizaremos como exemplo o K-Means.

De forma introdutória, é importante entendermos o funcionamento deste algoritmo, assim, separaremos esta exemplificação em uma sequência lógica que visa te auxiliar a compreensão de detalhes relacionados ao K-means como a execução do cálculo das distâncias, processo de avaliação do modelo e até mesmo o processo de escolha dos agrupamentos (número ideal de clusters). 

## O que é K-MEANS?

O **K-Means** é um algoritmo que trata preferencialmente de dados numéricos contínuos e que tem o objetivo de dividir as observações em k clusters, onde **K** é o número de clusters. Esta divisão dos dados entre os clusters busca atender duas principais premissas (ou ainda que não atinja, esse é o objetivo de dividir os dados):
- Os dados dentro de um cluster deverão ser o mais similares possíveis, quando comparados dentro do seu agrupamento
- E os dados de outros clusters deverão ser o mais diferentes possíveis, quando comparados os clusters entre si

Para que isso ocorra, a forma que o algoritmo se utiliza é a soma dos erros quadrados ou Sum of Squared Errors (SSE), cujo objetivo é minimizar a distância entre os pontos e seu centróide (o centróide representa o centro do cluster). Depois que o modelo é criado, o K-means quando é aplicado lembra a média de cada cluster, também chamada de **centróides** e, com isso, encontra o centróide mais próximo a cada novo dado. 

Se pudéssemos desenhar um fluxo de processo da clusterização com K-means, poderíamos abordar as seguintes etapas:
1. Inicialização: etapa inicial em que o algoritmo identifica a localização dos centróides (são geradas aleatoriamente)
2. Designação: a partir da inicialização, os **K** clusters são criados, associando cada observação ao seu centróide mais próximo
3. Atualização: a cada nova associação de observação, o centróide de cada cluster se torna uma nova média do cluster, repetindo esse processo várias vezes até convergir, ou seja, até que todas as observações estejam agrupadas em seus clusters e essas observações tenham a menor distância possível para o centróide do cluster em que foi alocada

![](https://stanford.edu/~cpiech/cs221/img/kmeansViz.png)

**Esboço visual de como ocorre o processo de atualização do centróide, conforme as observações são agrupadas em seus clusters**

```python
# Importando as bibliotecas necessárias

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Configuração de visualização de células e número de casas decimais para dados do tipo float/flutuante

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.options.display.float_format = '{:.2f}'.format
```

```python
# Leitura do dataset de vinhos

df = pd.read_csv('wine.csv')
print(f"Quantidade de Linhas:", df.shape[0])
print(f"Quantidade de Colunas:", df.shape[1])
df.head(1)
```

```python
# Para fins de facilitação dos entendimentos posteriores, renomearei os nomes das colunas (exceto pH e Id)

df = df.rename(columns={"fixed acidity": "acidez_fixa", "volatile acidity": "acidez_volatil", "citric acid": "acido_citrico", 
                        "residual sugar": "acucar_residual", "chlorides": "cloretos",
                        "free sulfur dioxide": "dioxido_enxofre_livre", "total sulfur dioxide": "dioxido_enxofre_total", 
                        "density": "densidade", "sulphates": "sulfatos", "alcohol": "alcool", "quality": "qualidade"})
df.head(1)
```

```python
# Busco aqui verificar se cada linha do DF representa um único elemento de amostra, não se repetindo em outras linhas

if df['Id'].duplicated().any():
    print("Existem Id duplicadas no DF")    
else:
    print("Todas Id's são únicos, aparecendo uma única linha no DF")
```

```python
# Verificar a existência de nulos

nulos = df.isnull().sum()
nulos[nulos > 0]
```

```python
# Verificar detalhes de cada indicador

df.info()
```

### Preparação dos Dados - Pré-modelagem K-Means

**Gerando uma cópia do Dataframe para modelagem**
- Retirei apenas a coluna 'Id' do dataframe de modelagem, uma vez que ele não se trata de um atributo, mas sim de um identificador

```python
df.columns
```

```python
df_cluster = df[['acidez_fixa', 'acidez_volatil', 'acido_citrico', 'acucar_residual',
                 'cloretos', 'dioxido_enxofre_livre', 'dioxido_enxofre_total',
                 'densidade', 'pH', 'sulfatos', 'alcool', 'qualidade']]
df_cluster.describe()
```

**Analisando a visualização da estatística descritiva das variáveis acima, percebe-se que as escalas estão bem diferentes, por exemplo:**
- valor máximo de cloretos = 0.61
- valor máximo de dioxido_enxofre_total = 289

Como estamos tratando com um algoritmo K-means que calcula a distância entre os pontos, essa distorção pode trazer ruídos indesejados ao algoritmo, desta forma, uma boa prática seria aplicar uma técnica para normalizar ou padronizar os dados.
- Para este exemplo, utilizaremos o padronizador **StandardScaler** da biblioteca Scikit-Learn

Usando StandardScaler() o dataframe distribuirá os valores das variáveis em uma mesma escala, diminuindo assim a variância dos dados, uma vez que ele fará com que a média das variáveis seja igual a 0 e o desvio padrão igual a 1.

```python
from sklearn.preprocessing import StandardScaler
```

```python
df_padronizado = StandardScaler().fit_transform(df_cluster) 
# o método .fit_transform() do StandardScaler retorna um array NumPy, e não um DataFrame do Pandas
# Por isso que na linha abaixo, precisamos realizar uma conversão para um DataFrame Pandas
df_padronizado = pd.DataFrame(df_padronizado, columns=df_cluster.columns)
df_padronizado.describe()
```

**Observemos que agora o problema das escalas foi solucionado**

### Identificação da quantidade de clusters

**Usando o método Elbow para identificar o número ideal de clusters?**

Um bom agrupamento deve ter poucos clusters e, ao mesmo tempo, manter a variação dos dados baixa. O Método do Cotovelo (Elbow Curve) ajuda a encontrar a quantidade ideal de clusters K. Esse método analisa como a variação dentro dos grupos diminui conforme o número de clusters aumenta. O ponto ideal é onde essa redução deixa de ser tão significativa, formando um “cotovelo” no gráfico. Esse é o melhor valor de K, pois mais clusters além desse ponto não trazem tanto benefício, apenas complexidade desnecessária.

![](https://dataat.github.io/introducao-ao-machine-learning/assets/04_clustering/metodo_cotovelo.png)

```python
elbow = []
K = range(2,21) # ponto de parada pode ser parametrizado manualmente
for k in K:
    kmeanElbow = KMeans(n_clusters=k, init='random', n_init='auto', random_state=42).fit(df_padronizado)
    elbow.append(kmeanElbow.inertia_)
    
plt.figure(figsize=(16,8))
plt.plot(K, elbow, marker='o')
plt.xlabel('Nº Clusters', fontsize=16)
plt.xticks(range(2,21))
plt.ylabel('WCSS', fontsize=16)
plt.title('Método de Elbow', fontsize=16)
plt.show()
```

### Silhueta

A silhueta é uma medida do quanto observações mais similares estão próximas entre si e, ao mesmo tempo, o quanto estão distantes de outros clusters diferentes. Busca maior **coesão** dentro do cluster e maior **separação** entre clusters. A silhueta vai de -1 a 1, onde o maior valor indica que a observação se encaixou mais dentro do cluster e se distanciou dos outros clusters. Se a silhueta for negativa, a configuração do cluster pode ter ou clusters de mais ou clusters de menos. A silhueta também usa a distância euclidiana para medir a distância entre as observações.
![](http://www.mtechprojects.org/wp-content/uploads/2017/12/silhouette-coefficient.jpg)

Por exemplo:
![](https://www.researchgate.net/profile/Frans_Coenen/publication/221570710/figure/fig1/AS:670029003644935@1536758771429/Derivation-of-the-Overall-Silhouette-Coefficient-OverallSil.png)

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []
K = range(2, 21)  # Precisei alterar o range, pois a silhueta não é definida para k=1

for k in K:
    kmeans = KMeans(n_clusters=k, init='random', n_init='auto', random_state=42).fit(df_padronizado)
    labels = kmeans.labels_
    silhouette_avg = silhouette_score(df_padronizado, labels)
    silhouette_scores.append(silhouette_avg)

# Aqui eu peço para que seja destacado o pico mais alto, que aponta o numéro ÓTIMO de clusters
best_k = K[silhouette_scores.index(max(silhouette_scores))]
best_score = max(silhouette_scores)

# Plotar os resultados
plt.figure(figsize=(16, 8))
plt.plot(K, silhouette_scores, marker='o', label='Coeficiente de Silhueta')
plt.axvline(x=best_k, linestyle='--', color='red', label=f'Melhor k: {best_k}')
plt.scatter(best_k, best_score, color='red')  # Destacar o ponto do pico
plt.xlabel('Nº Clusters', fontsize=16)
plt.xticks(range(2, 21))
plt.ylabel('Média do Coeficiente de Silhueta', fontsize=16)
plt.title('Análise de Silhueta para Identificação do Nº Ótimo de Clusters', fontsize=16)
plt.legend()
plt.show()
```

```python
# Aplicar K-Means
kmeans = KMeans(n_clusters=best_k, init='random', n_init=10, random_state=42)
clusters = kmeans.fit_predict(df_padronizado)
```

```python
kmeans_clusters = kmeans.labels_
df['cluster_kmeans'] = kmeans_clusters
df['cluster_kmeans'] = df['cluster_kmeans'].astype('category')
```

```python
kmeans_clusters
```

```python
cent_finais = pd.DataFrame(kmeans.cluster_centers_)
cent_finais.columns = df_padronizado.columns
cent_finais.index.name = 'cluster'
cent_finais
```

```python
print(df['cluster_kmeans'].value_counts(normalize=True))
df['cluster_kmeans'].value_counts()
```

```python
from sklearn.decomposition import PCA

# Aplicar PCA nos dados
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_padronizado)

# Criar DataFrame para visualização
df_pca = pd.DataFrame(df_pca, columns=['PC1', 'PC2'])
df_pca['Cluster'] = clusters

# Plotar os clusters
plt.figure(figsize=(8,6))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', palette='tab10', data=df_pca, alpha=0.7, edgecolor='k')
centroids_pca = pca.transform(pd.DataFrame(kmeans.cluster_centers_, columns=df_padronizado.columns))
plt.scatter(centroids_pca[:, 0], centroids_pca[:, 1], marker='X', s=200, c='red', label='Centroides')
plt.title(f'Distribuição dos Clusters (K={best_k})')
plt.legend()
plt.show()
```

```python
print(f"Melhor índice de silhueta apresentou {best_k} clusters com a média do coeficiente de silhueta = {round(best_score,4)}")
```

Como visto na explicação do conceito do método de silhueta, que varia de -1 a 1, em que valores próximos de 1 indicam que a observação está bem dentro do seu cluster, enquanto valores negativos sugerem que a clusterização pode não estar ideal, entendemos que a distribuição não está boa, assim, partiremos para uma nova etapa de preparação dos dados.

A partir dos resultados acima, entende-se ser uma alternativa testar a criação de novos indicadores, conforme a sequência abaixo

### **acidez_fixa_ph_ratio** - Proporção entre acidez fixa e pH (Indica equilíbrio ácido do vinho)

```python
df_cluster.loc[:, 'acidez_fixa_ph_ratio'] = df_cluster['acidez_fixa'] / df_cluster['pH']
```

### **alcool_densidade_ratio** - Razão entre álcool e densidade (Indica concentração alcoólica do vinho)

```python
df_cluster.loc[:, 'alcool_densidade_ratio'] = df_cluster['alcool'] / df_cluster['densidade']
```

### **acidez_total** - Composição ácida combinada (Soma da acidez fixa e ácido cítrico, reflete acidez total)

```python
df_cluster.loc[:, 'acidez_total'] = df_cluster['acidez_fixa'] + df_cluster['acido_citrico']
```

### **indice_equilibrio_acido** - Índice de equilíbrio ácido-volátil (Indica a relação entre acidez cítrica e volátil, útil para avaliar o equilíbrio do vinho)

```python
df_cluster.loc[:, 'indice_equilibrio_acido'] = df_cluster['acido_citrico'] / (df_cluster['acidez_volatil'] + 1e-6)  # Pequeno valor para evitar divisão por zero
```

### **acidez_fixa_ph_ratio** - Proporção entre acidez fixa e pH (Indica equilíbrio ácido do vinho)

```python
df_cluster.loc[:, 'so2_ratio'] = df_cluster['dioxido_enxofre_livre'] / (df_cluster['dioxido_enxofre_total'] + 1e-6)
```

### **impacto_alcool_qualidade** - Índice de impacto alcoólico na qualidade (Multiplica álcool pela qualidade, enfatizando sua influência no vinho)

```python
df_cluster.loc[:, 'impacto_alcool_qualidade'] = df_cluster['alcool'] * df_cluster['qualidade']
```

### **qualidade_categoria - Criando níveis de qualidade para os vinhos**

```python
df_cluster['qualidade'].value_counts()
```

```python
def categorizar_qualidade(q):
    if q <= 4:
        return 0 # (qualidade mais baixa)
    elif 5 <= q <= 6:
        return 1 # (qualidade média/equilibrada)
    else:
        return 2 # (qualidade mais alta)
    
# Aplicar função ao dataframe
df_cluster['qualidade_categoria'] = df_cluster['qualidade'].apply(categorizar_qualidade)
```

```python
elbow = []
K = range(2,21) # ponto de parada pode ser parametrizado manualmente
for k in K:
    kmeanElbow = KMeans(n_clusters=k, init='random', n_init='auto', random_state=42).fit(df_padronizado)
    elbow.append(kmeanElbow.inertia_)
    
plt.figure(figsize=(16,8))
plt.plot(K, elbow, marker='o')
plt.xlabel('Nº Clusters', fontsize=16)
plt.xticks(range(2,21))
plt.ylabel('WCSS', fontsize=16)
plt.title('Método de Elbow', fontsize=16)
plt.show()
```

```python
silhouette_scores = []
K = range(2, 21)  # Precisei alterar o range, pois a silhueta não é definida para k=1

for k in K:
    kmeans = KMeans(n_clusters=k, init='random', n_init='auto', random_state=42).fit(df_padronizado)
    labels = kmeans.labels_
    silhouette_avg = silhouette_score(df_padronizado, labels)
    silhouette_scores.append(silhouette_avg)

# Aqui eu peço para que seja destacado o pico mais alto, que aponta o numéro ÓTIMO de clusters
best_k = K[silhouette_scores.index(max(silhouette_scores))]
best_score = max(silhouette_scores)

# Plotar os resultados
plt.figure(figsize=(16, 8))
plt.plot(K, silhouette_scores, marker='o', label='Coeficiente de Silhueta')
plt.axvline(x=best_k, linestyle='--', color='red', label=f'Melhor k: {best_k}')
plt.scatter(best_k, best_score, color='red')  # Destacar o ponto do pico
plt.xlabel('Nº Clusters', fontsize=16)
plt.xticks(range(2, 21))
plt.ylabel('Média do Coeficiente de Silhueta', fontsize=16)
plt.title('Análise de Silhueta para Identificação do Nº Ótimo de Clusters', fontsize=16)
plt.legend()
plt.show()

print(f"Melhor índice de silhueta apresentou {best_k} clusters com a média do coeficiente de silhueta = {round(best_score,4)}")
```
