## **5Q's**
Quais são os dados de entrada? \
O que eu vou fazer com eles? \
Quais são as restrições? \
Qual o resultado esperado? \
Quais os passos eu vou seguir? 

### **Exemplo: Chute um Número** 
**Quais os dados de entrada?** \
Valor do número chutado pelo usuário \
Valor aleatório de 1 a 10 a ser gerado após o iníco do programa

**O que eu vou fazer com eles?** \
Após receber o valor do usuário, devo pedir um valor aleatório de 1 a 10 e validar se o usuário acertou, se ficou acima ou abaixo, para depois informar ao usuário o resultado \
Se ele acertar interrompe o programa, se ele errar, jogar novamente até acertar

**Quais são as restrições?** \
Caso o usuário digite um número maior que 10 ou menor que 1, devo emitir uma mensagem pedindo um outro número que esteja no intervalo de 1 a 10 e não iniciar o programa enquanto isso \
O programa não deve ser finalizado até que um numero seja acertado e o usuário deve jogar quantas vezes quiser, até acertar

**Qual o resultado esperado?** \
O programa identificar que o valor chutado pelo usuário é igual ao gerado no início do programa

**Quais os passos eu vou seguir?** \
Gerar o valor aleatório de 1 a 10 \
Coletar o valor do usuário \
Se ele mandar um valor fora do intervalo, enviar uma mensagem e só prosseguir após ele informar um numero dentro do intervalo \
Se o valor do usuário for menor que o numero gerado, informar que o valor está abaixo e disponibilizar uma nova chance de rodar \
Se o valor do usuário for maior que o numero gerado, informar que o valor está acima e disponibilizar uma nova chance de rodar \
Se o valor do usuário for igual, informar que o usuário acertou e encerrar o programa

## Pseudocódigo: 
Criar uma solução de forma escrita, descrevendo os passos em linguagem natural - algoritmo \
É apenas criado para ser uma lógica que resolve um problema, depois disso aplicar em uma linguagem de programação \

**SINTAXE:**
- Input
- Print
- If e Else (condição)
- Loop:
    - X a Y (iterando do X até o Y)
    - X em Y (iterando do X em um conjunto Y)
    - While X (enquanto X for verdadeiro)

Resultado

1fat = 5   (i * fat1 = 5 * 1) \
