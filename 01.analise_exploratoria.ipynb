## Limpeza e Pré-Processamento dos Dados

- Como lidar com valores ausentes? Depende do contexto, que sempre deve ser ponderado e analisado:

    - Por exemplo, quando estamos falando que temos um dataset em que as variáveis são do tipo numéricas contínuas (como altura ou peso de uma pessoa), pode ser conveniente substituir os dados ausentes pela média obtida sobre a amostra analisada
    - No entanto, imagine fazer o mesmo com a quantidade de filhos que uma família possuí. Talvez nesse caso, o valor nulo seja melhor preenchido com zero (0)
    - E se estamos tratando com variáveis categóricas, uma estratégia seria considerar a moda (valores que mais se repetem)
    - Tudo dependerá de uma análise minuciosa do negócio, problema apresentado e impacto que a substituição terá para as análises e/ou resultados esperados

```python
# Em nosso caso, não há a presença de valores nulos, portanto, o código abaixo tem a função de exemplificar

# Substituição dos valores ausentes por 0
df['coluna'] = df['coluna'].fillna(0)

# Substituição dos valores ausentes pela média
df["coluna"] = df["coluna"].fillna(df["coluna"].mean())

# Remover todas as linhas com valores ausentes para uma coluna específica
# df = df.dropna(subset=['coluna']) 'ou' df.dropna(axis=0, inplace=True)

# Aqui está um exemplo de como remover todas as linhas com valores ausentes 
df = df.dropna()
# apesar de ser uma medida mais dura, em determinados casos se torna mais conservador e assertivo seguir desta forma
```

## Análise de Distribuições (Histogramas)
- A intenção nesta fase é compreender a distribuição dos dados, por isso, utilizaremos um gráfico do tipo Histograma, que possuí uma visualização e interpretação mais facilitada

```python
# Exemplo: histograma para distribuição de uma variável numérica

coluna = df["volatile acidity"]
sns.histplot(coluna, kde=True, bins=30)
plt.title("Histograma")
plt.show()
```

## Análise de Dispersão (verificação de outliers)
- Nesta etapa, estamos interessados em visualizar como estão as distribuições dos dados na variável, compreender a dispersão deles em relação as medidas centrais, como média e mediana, bem como notar a existência de outliers

    - **Média:** É influenciada por valores extremos (outliers)
    - **Mediana:** É mais robusta para distribuições com outliers

```python
# Boxplot para identificar outliers
sns.boxplot(x=df["volatile acidity"])
plt.title("Outliers na variável")
plt.show()
```

```python
#Relações Entre Variáveis com gráficos de dispersão
sns.scatterplot(x='quality', y='alcohol', data=df)
plt.show()
```

```python
# Gráfico de Barras: Contagem de cada categoria
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='quality', order=df['quality'].value_counts().index)
plt.title('Distribuição de Variáveis Categóricas', fontsize=14)
plt.xlabel('Quality', fontsize=12)
plt.ylabel('Frequência', fontsize=12)
plt.show()
```

## Análise de Correlação

- Aqui, busca-se criar uma matriz de correlação, que demonstra o coeficiente de correlação (variando de -1 até 1)
- A intenção é buscar entender se existe correlação entre um par de variáveis e a partir disso, interpretar o impacto para o negócio/modelo

```python
# Aqui criamos a matriz de correlação com o dataset atual
correlacao = df.corr()

# Depois disso, crio um heatmap da correlação, que seria a plotagem dessa matriz
plt.figure(figsize=(12, 8))
sns.heatmap(correlacao, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlação entre variáveis")
plt.show()
```
